Assim como na Lógica Proposicional, a Lógica de Primeira-Ordem pode ser utilizada para representar (parte) do mundo, ou seja, podemos especificar fatos acerca do mundo na linguagem lógica. Podemos utilizar a lógica para "raciocinar" através de sua relação de consequência lógica ($\models$) ou de derivação ($\vdash$).
  
image.png

Na lógica de primeira-ordem, utilizamos os símbolos $\exists$ e $\forall$ para quantificar propriedades existenciais e universais sobre conjuntos, tais como:

- "Existe um número par" $\rightsquigarrow$  $\exists x~Par(x)$.
- "Existe alguém que ama Maria" $\rightsquigarrow$ $\exists x~Ama(x, Maria)$.
- "Existe alguém que trabalha e ama Maria" $\rightsquigarrow\exists x~Trabalha(x)\wedge Ama(x, Maria)$.
- "Todos os números são maiores do que zero" $\rightsquigarrow$  $\forall x~ x\geq 0$.
- "Todos os números são maiores do que zero" $\rightsquigarrow$  $\forall x~ x\geq 0$.
- "Existe um número que é menor do que todos os outros" $\rightsquigarrow\exists x\forall y~ x\leq y$.

## Relações Frequentes
- Duas **relações são disjuntas**, e a asserção de uma implica a negação da outra, como em:
$$\forall x(Homem(x)\rightarrow \lnot Mulher(x))$$

- Existem muitas relações (predicados) que implicam em alguma forma de **especialização (subtipo)**, na qual um tipo está subsumido no outro, e.g.,
$$\forall x(Cirurgiao(x)\rightarrow Medico(x))$$

- Quando dois ou mais subtipos representam completamente um **supertipo**, como em:
  $$\forall x(Adulto(x)\rightarrow (Homem(x)\vee Mulher(x)))$$
- Algumas relações são **simétricas**:
$$\forall x\forall y(Casado(x,y)\rightarrow Casado(y,x))$$
- Algumas relações são opostas uma da outra (relação **inversa**):
$$\forall x\forall y(FilhoDe(x,y)\rightarrow  PaiDe(y,x))$$ 
- **Parte** do significado de algum predicado é o fato de seus argumentos serem um certo tipo:
$$\forall x\forall y(Casado(x,y)\rightarrow (Pessoa(x)\wedge Pessoa(y)))$$
- Quando queremos criar predicados **compostos** que são completamente definidos por uma combinação de outros predicados:
$$\forall x(HomemRico(x)\leftrightarrow (Homem(x)\wedge Rico(x)))$$
- A **negação de uma quantificação universal**, e.g. "nem todos são ricos" significa que "existem indivíduos que não são ricos":
$$\lnot \forall x~Rico(x)\leftrightarrow \exists x \lnot Rico(x)$$
- A **negação de uma quantificação existencial**, e.g. "não existe unicórnio" significa que "todos não são unicórnios":
$$\lnot \exists x~Unicornio(x)\leftrightarrow \forall x \lnot Unicornio(x)$$

## Especificação do Problema do Clube
**(Problema do Clube)** Ana, Maria e João pertencem ao Balneário Clube. Todo o membro do Balneário que não é um nadador faz rapel. Quem faz rapel não gosta de chuva e qualquer um que não gosta de mergulhar não é um nadador. Maria não gosta de qualquer coisa que João goste e gosta de tudo que João não gosta. João gosta de chuva e de mergulhar. Existe alguém que gosta de rapel, mas não de nadar?

> **Resposta:** 
>
> - "Ana, Maria e João pertencem ao Balneário Clube"
    $$BalnearioClube(ana)$$
    $$BalnearioClube(maria)$$ 
    $$BalnearioClube(joao)$$
- "Todo o membro do Balneário que não é um nadador faz rapel."
$$\forall x((BalnearioClube(x)\wedge\lnot FazNatacao(x))\rightarrow FazRapel(x))$$
- "Quem faz rapel não gosta de chuva"
$$\forall x(FazRapel(x)\rightarrow \lnot GostaChuva(x))$$
- "qualquer um que não gosta de mergulhar não é um nadador"
$$\forall x(\lnot GostaMergulhar(x)\rightarrow \lnot FazNatacao(x))$$    
- "Maria não gosta de qualquer coisa que João goste e gosta de tudo que João não gosta"
    $$GostaMergulhar(joao)\rightarrow\lnot GostaMergulhar(maria)$$
    $$GostaChuva(joao)\rightarrow\lnot GostaChuva(maria)$$
    $$\lnot GostaMergulhar(joao)\rightarrow GostaMergulhar(maria)$$
    $$\lnot GostaChuva(joao)\rightarrow\ GostaChuva(maria)$$    
- "João gosta de chuva e de mergulhar"
    $$GostaChuva(joao)~~GostaMergulhar(joao)$$
- "Existe alguém que gosta de rapel, mas não de nadar"
    $$\exists x(FazRapel(x)\wedge \lnot FazNatacao(x)) $$
