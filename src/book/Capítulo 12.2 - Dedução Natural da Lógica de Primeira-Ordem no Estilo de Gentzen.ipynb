## Restrições nas Regras 
É importante observar as retrições das regras de inferência de introdução e eliminação dos quatificadores. Caso contrário, podemos fazer demostrações erradas. A seguir apresentamos algumas utilizações erradas das regras de introdução e eliminação dos quatificadores que não respeitam as restrições e, portanto, geram provas erradas!

- Sem a restrição “a não pode ocorrer em $\varphi$” e de $\exists x\varphi(x)$, é possível deduzir
**incorretamente** que: $\forall x\varphi(x)$
    
    image.png

- Sem a restrição "a não pode ocorrer em $\exists x\varphi(x)$" e de $\forall y\exists x(y<x)$, é possível deduzir **incorretamente** que: $\exists x(x<x)$

    image.png

- Sem a restrição "a não pode ocorrer em qualquer hipótese em que a dedução de $\alpha$ dependa" e de $\exists x~Par(x),~ \exists x~Impar(x)$, é possível deduzir **incorretamente** que: $\exists x(Par(x)\wedge Impar(x))$

    image.png

- Prove que $\forall x(H(x)\rightarrow M(x)), \forall x \lnot M(x)\vdash \forall x\lnot H(x)$

    image.png

- Prove que $\forall x\varphi(x),\forall x\psi(x)\vdash \forall x(\varphi(x)\wedge\psi(x))$ 

    image.png

- Prove que $\exists x(\varphi(x)\wedge\psi(x))\vdash \exists x\varphi(x)\wedge\exists x\psi(x)$
