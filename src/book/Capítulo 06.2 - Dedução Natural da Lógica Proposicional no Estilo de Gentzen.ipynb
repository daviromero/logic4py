# [Dedução Natural no Estilo de Gentzen](https://youtu.be/BQUXLaeCaNk)
No estilo de Gentzen, as provas são apresentadas na forma de árvores onde a raiz da árvore é a fórmula que se quer provar e as folhas representam as premissas ou suposições que serão descartadas. Os nós intermediários são aplicações de regras de derivação que podem ser de introdução ou eliminação de conectivos. As folhas da árvore são as premissas ou hipóteses que foram introduzidas por regras de derivação e que serão posteriormente descartadas. As regras de derivação unem árvores finitas, que são geradas a partir de um conjunto finito de premissas e hipóteses até derivar uma certa conclusão. Uma fórmula introduzida como hipótese será representada entre chaves [ ] e numerada, onde o número será usado para indicar o descarte dessa hipótese por uma regra de derivação em algum passo posterior. Uma hipótese descartada não pode ser utilizada após a regra que introduziu a hipótese. 



## Regras da Conjunção 

A regra da **conjunção introdução ($\land i$)** é apresentada na figura abaixo e diz que se temos uma demonstração de $\varphi$ e outra demonstração de $\psi$, então podemos compor essas demonstrações e concluir $\varphi\land\psi$.

image.png

A regra da **conjunção eliminação ($\land e$)** é apresentada na figura abaixo e diz que se temos uma demonstração de $\varphi\land\psi$, podemos construir uma demonstração de $\varphi$ (ou $\psi$).

image.png

A regra da **implicação eliminação ($\rightarrow e$)** é apresentada na figura abaixo e diz que se temos uma demonstração de $\varphi\rightarrow\psi$ e outra demonstração de $\varphi$, podemos construir uma demonstração de $\psi$.

image.png

A regra da **implicação introdução ($\rightarrow i$)** é apresentada na figura abaixo e diz que se fizermos a suposição de $\varphi$ e construirmos uma demonstração de $\psi$, podemos concluir $\varphi\rightarrow\psi$. Neste caso $\varphi$ é uma suposição temporária que será assinalada entre $[$ e $]$ e conterá um número $n$ que irá indicar a regra da $\rightarrow i$ que a introduziu. Na árvore de derivação abaixo da demonstração de $\varphi\rightarrow\psi$ não poderemos mais usar $[\varphi]^n$, pois ele foi descartado.

image.png


A regra da **disjunção introdução ($\lor i$)** é apresentada na figura abaixo e diz que temos uma demonstração de $\psi$ (ou $\psi$), podemos concluir $\varphi\lor\psi$. 

image.png


A regra da **disjunção eliminação ($\lor e$)** é apresentada na figura abaixo e diz que podemos concluir $\alpha$ se temos as seguintes demonstrações:
1. Se temos uma demonstração de $\varphi\lor\psi$;
1. Se fizermos uma suposição temporária de $[\varphi]^n$ e demonstrarmos $\alpha$; e
1. Se fizermos uma suposição temporária de $[\psi]^m$ e demonstrarmos $\alpha$.

Nesses casos iremos descartar as hipóteses $[\varphi]^n$ e $[\psi]^m$ após a eliminação da disjunção. É importante frisar que a hipótese $[\varphi]^n$ (ou $[\psi]^m$) não pode ser utilizada na demonstração de $\alpha$ que supõe $[\psi]^m$ (ou $[\varphi]^n$).

image.png

A regra da **negação introdução ($\lnot i$)** é apresentada na figura abaixo e diz que se fizermos a suposição de $\varphi$ e construirmos uma demonstração da contradição $\bot$, podemos concluir $\lnot\varphi$. Note que esta regra é semelhante a regra da implicação introdução, na qual podemos assumir $\lnot\varphi\equiv\varphi\rightarrow\bot$.

image.png


A regra da **negação eliminação ($\lnot e$)** é apresentada na figura abaixo e diz que se temos uma demonstração de $\varphi$ e outra demonstração de $\lnot\varphi$, podemos concluir a contradição $\bot$.

image.png

A regra da **contradição eliminação** ($\bot e$) é apresentada na figura abaixo e diz que se demonstrarmos uma contradição, então podemos construir que uma demonstração de uma fórmula qualquer $\varphi$.

image.png

A regra de **redução ao absurdo** ($raa$) é apresentada abaixo e nos diz que demonstrarmos a contradição $\bot$ a a partir da suposição de $\lnot\varphi$, então podemos construir a demonstração de $\varphi$.

image.png
