# Introdução

No início do século passado, correntes filosóficas surgiram objetivando a fundamentação da matemática em bases sólidas utilizando a lógica para tal. A escola formalista de Hilbert, por exemplo, buscava a prova da auto-consistência da matemática. Em particular, ele buscava solucionar este problema de uma forma automatizada através de um procedimento efetivo (mecânico) que fosse capaz de verificar a consistência das proposições matemáticas. Apesar dos resultados negativos de Gödel quanto à impossibilidade uma formalização completa da matemática e da sua prova de auto-consistência publicados no início da década de 30, a escola formalista seguiu buscando uma melhor caracterização do conceito de procedimento efetivo capaz de decidir propriedades de fórmulas. Os primeiros métodos automatizáveis para a prova de teoremas foram introduzidos por Herbrand ainda nesta década. Contudo, apenas com o desenvolvimento dos computadores e o interesse surgido pela automatização do raciocínio em Inteligência Artificial na década de 60 é que métodos automatizáveis de prova mais eficientes, como o sistema de resolução foram desenvolvidos.

Como a escola formalista buscava a prova da auto-consistência da matemática, provas em matemática passaram a ser objetos de estudo da própria matemática dando início a uma área conhecida como Teoria da Prova. Um dos conceitos centrais desta teoria é o de sistema dedutivo que pode ser entendido como um mecanismo que permite a construção de argumentos formais, estabelecendo conclusões a partir de premissas.

O sistema dedutível de Dedução Natural surgiu neste contexto buscando uma melhor análise das estrutura das provas lógicas. Vários resultados teóricos foram alcançados como aquele que afiram a existência de provas normais, diretas, para provas em lógica (clássica e intuicionística). Como consequência, demonstrou-se que a lógica era auto-consistente sem a necessidade de apelar para outros modelos.

O sistema de Dedução Natural é uma ferramenta essencial para a Teoria da Prova, pois permitem um maior entendimento da noção de prova e do papel dedutivo das constantes lógicas, individualizadas pelas regras de introdução e eliminação. Nos sistemas de Dedução Natural, as inferências são quebradas em passos atômicos de tal forma que cada passo envolve uma constante lógica diferente. Tais passos são de dois tipos: aqueles que permitem a introdução do signo lógico; e, os que eliminam tais signos. As provas inciam a partir de premissas que são assumidas ou de suposições que são necessariamente descartadas no decorrer da prova. As derivações são, em geral, apresentadas na forma de árvores, também conhecido como estilo de Gentzen, ou por meio de caixas, denominado de estilo de Fitch. 

O sistema de Dedução Natural é amplamente utilizados para o ensino de demonstrações e consta de muitos dos livros-texto de Lógica. O artigo "*A Brief History of Natural Deduction*" apresenta uma breve história de Dedução Natural, traçando o desenvolvimento da dedução natural do ponto de vista
que esses fundadores abraçaram à ampla aceitação do método na década de 1960. Apresenta as diferentes escolhas feitas por escritores de livros-texto com o objetivo de determinar quais são as "características essenciais" da natureza dedução são. O artigo "*A History of Natural Deduction and Elementary Logic Textbooks*" compara o estilo de demonstrações de Dedução Natural de 33 livros-texto, no qual o estilo de Fitch é adotado por 19 destes e o de Gentzen por 4.


No estilo de Gentzen, as provas são apresentadas na forma de árvores onde a raiz da árvore é a fórmula que se quer provar e as folhas representam as premissas ou suposições que serão descartadas. Os nós intermediários são aplicações de regras de derivação que podem ser de introdução ou eliminação de conectivos. As folhas da árvore são as premissas ou hipóteses que foram introduzidas por regras de derivação e que serão posteriormente descartadas. As regras de derivação unem árvores finitas, que são geradas a partir de um conjunto finito de premissas e hipóteses até derivar uma certa conclusão. Uma fórmula introduzida como hipótese será representada entre chaves [ ] e numerada, onde o número será usado para indicar o descarte dessa hipótese por uma regra de derivação em algum passo posterior. Uma hipótese descartada não pode ser utilizada após a regra que introduziu a hipótese. 

A Figura 1a apresenta a prova da transitividade, na qual $A\rightarrow B$ e $B\rightarrow C$ são assumidas como premissas (folhas da árvore) e $A\rightarrow C$ é a conclusão (raiz da árvore). A regra de introdução da implicação ($\rightarrow i$) é utilizada para demonstrar $A\rightarrow C$, onde a fórmula $A$ é assumida (folha $\big[A\big]^{_{1}}$ que é descartada nesta regra) para concluir $C$. Para demonstrar $C$ (nó intermediário), a regra de inferência eliminação de implicação ($\rightarrow e$), também conhecida como *Modus Ponens*, é utilizada a partir da premissa $B\rightarrow C$ e da prova de $B$. Por fim, para demonstrar $B$ (nó intermediário), a regra de *Modus Ponens* é aplicada novamente a partir da hipótese $A$ e da premissa $A\rightarrow B$.

No estilo de Fitch, as demonstrações são apresentadas de forma linear e sequencial, na qual cada uma das linhas da prova é numerada, tem uma afirmação e uma justificativa. As justificativas são definidas por serem premissas da prova ou pela aplicação de uma das regras do sistema dedutível de Dedução Natural. As subprovas dentro de uma prova maior têm caixas ao redor e servem para delimitar o escopo de hipóteses temporárias. Provas podem ter caixas dentro de caixas, ou pode-se abrir outras caixas depois de fechar outras, obedecendo as regras de demonstração. Uma fórmula só pode ser utilizada em uma prova em um determinado ponto se essa fórmula aconteceu anteriormente e se nenhuma caixa que contenha essa ocorrência da fórmula tenha sido fechada.

A Figura 1b apresenta a derivação da transitividade, na qual as fórmulas $A\rightarrow B$ e $B\rightarrow C$ são assumidas como premissas nas linhas $2$ e $3$, respectivamente, e a fórmula $A\rightarrow C$ é concluída na linha $6$ (última linha), por meio da regra da introdução da implicação ($\rightarrow i$) das linhas 3 até 5, que formam uma caixa, onde a fórmula $A$ é assumida como hipótese (primeira linha da caixa) na linha $3$ e a fórmula $C$ é concluída na linha $5$ (última linha da caixa). 
A fórmula $B$ é demonstrada na linha $4$ a partir da regra da eliminação de implicação pelas linhas $3$ ($A$) e $1$ ($A\rightarrow B$). Por fim, $C$ é provado na linha $5$ pela regra da eliminação da implicação aplicada as linhas $4$ ($B$) e $2$ ($B\rightarrow C$).
